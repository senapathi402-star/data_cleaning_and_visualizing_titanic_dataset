# **Titanic_data_cleaning**

## **Titanic**

The RMS Titanic was a massive, luxurious British ocean liner. Built by Harland and Wolff, it was considered "practically unsinkable" due to advanced features like watertight compartments.

image.png

```python
import seaborn as sn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = sn.load_dataset('titanic')
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df['embarked'].unique()
```

```python
df['embarked'].fillna(df['embarked'].mode()[0],inplace = True)
```

```python
df['embark_town'].unique()
```

```python
df['embark_town'].fillna(df['embark_town'].mode()[0],inplace = True)
```

```python
df.isna().sum()
```

```python
df.groupby('sex')['age'].mean().round(0)
```

```python
df['age'].fillna(df.groupby('sex')['age'].mean()[0].round(0),inplace = True)
```

```python
df.isna().sum()
```

```python
df['deck'].unique()
```

```python
df['deck'] = df['deck'].fillna(df.groupby('pclass')['deck'].transform(lambda x: x.mode()[0] if not x.mode().empty else np.nan))
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df['age'] = df['age'].astype(int)
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df.head()
```

# Data Visualization

##Univariate analysis

###Numerical

```python
df['pclass'].value_counts()
```

```python
plt.figure(figsize = (4,2))
sns.countplot(data = df,y = 'pclass')
```

###Categorical

```python
df['deck'].value_counts()
```

```python
plt.figure(figsize = (6,3))
sns.countplot(data = df,x = 'deck')
plt.title('passengers on different deck')
plt.show()
```

##Bivatiate analysis

###Categorical - Numerical

```python
plt.figure(figsize = (6,3))
sns.countplot(data = df,x = 'deck',hue = 'survived')
plt.title('passengers on different deck')
plt.legend(loc = 9,bbox_to_anchor= (0.5,1.25),ncol = 2)
plt.show()
```

###Categorical - Categorical

```python
data = df.groupby('sex')['alive'].value_counts()
data
```

```python
plt.figure(figsize = (6,2))
plt.pie(data,labels = data.index,autopct = '%1.2f%%')
plt.title('Percentage of male and female survivers')
plt.show()
```
